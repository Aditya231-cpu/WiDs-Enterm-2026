```python
import requests
import pandas as pd
import numpy as np
import yfinance as yf
from bs4 import BeautifulSoup
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from transformers import pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fetch news using yfinance
ticker = yf.Ticker("AAPL")
yfinance_news = ticker.news

news = []
for article in yfinance_news:
    # Accessing nested 'title' and 'displayTime' from the 'content' dictionary
    if 'content' in article and 'title' in article['content'] and 'displayTime' in article['content']:
        news.append({
            "title": article["content"]["title"],
            "published": pd.to_datetime(article["content"]["displayTime"]) # Convert string datetime to pandas datetime
        })

news_df = pd.DataFrame(news, columns=["title", "published"])

# Proceed with date conversion and processing as before
if news_df.empty:
    print("Warning: No news items found or parsed successfully. Initializing an empty DataFrame with expected columns and types.")
    news_df["published"] = pd.Series(dtype='datetime64[ns]')
    news_df["date"] = pd.Series(dtype='object')
else:
    # Ensure 'published' is datetime and extract 'date'
    news_df["published"] = pd.to_datetime(news_df["published"])
    news_df["date"] = news_df["published"].dt.date

news_df.head()
```

```python
# =============== VADER Sentiment ==============

analyzer = SentimentIntensityAnalyzer()

news_df["vader"] = news_df["title"].apply(
    lambda x: analyzer.polarity_scores(x)["compound"]
)

news_df.head()
```

```python
# =============== FinBERT Sentiment =============

print("Loading FinBERT (first run slow)...")

finbert = pipeline("sentiment-analysis", model="ProsusAI/finbert")

def finbert_score(text):
    result = finbert(text[:512])[0]
    if result["label"] == "positive":
        return result["score"]
    elif result["label"] == "negative":
        return -result["score"]
    else:
        return 0

news_df["finbert"] = news_df["title"].apply(finbert_score)

news_df.head()
```

```python
# ========== Daily Sentiment Aggregation =========

daily_sentiment = news_df.groupby("date").agg({
    "vader": ["mean", "std", "count"],
    "finbert": ["mean"]
})

daily_sentiment.columns = [
    "vader_mean",
    "vader_std",
    "news_count",
    "finbert_mean"
]

daily_sentiment = daily_sentiment.reset_index()

daily_sentiment.head()
```

```python
# ============== Download AAPL Prices =============

# Adjusting the start date to align with the available news data more closely
price_df = yf.download("AAPL", start="2025-12-01") # Changed start date to 2025-12-01
price_df = price_df.reset_index()
price_df["date"] = price_df["Date"].dt.date

price_df["return"] = price_df["Close"].pct_change()
price_df["target"] = (price_df["return"].shift(-1) > 0).astype(int)

price_df = price_df.dropna()

price_df.head()
```

```python
# ============== Merge News + Prices =============

# Flatten MultiIndex columns of price_df before merging
# This converts columns like ('Price', 'Close') into 'Price_Close'
price_df.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in price_df.columns.values]

# Rename the relevant columns to match the expected single-level names
# Corrected: 'Date_' should be mapped to 'Date'
price_df = price_df.rename(columns={'Date_': 'Date', 'Close_AAPL': 'Close', 'High_AAPL': 'High', 'Low_AAPL': 'Low', 'Open_AAPL': 'Open', 'Volume_AAPL': 'Volume'})

# Re-create the 'date', 'return', and 'target' columns with the flattened column names
price_df["date"] = pd.to_datetime(price_df["Date"]).dt.date
price_df["return"] = price_df["Close"].pct_change()
price_df["target"] = (price_df["return"].shift(-1) > 0).astype(int)

# Re-apply dropna after changes if any NaNs were introduced or for consistency
price_df = price_df.dropna()

data = pd.merge(price_df, daily_sentiment, on="date", how="left")
data = data.fillna(0)

data.head()
```

```python
# ================= Features ====================

features = [
    "vader_mean",
    "vader_std",
    "news_count",
    "finbert_mean"
    # "return" # Temporarily removed to observe sentiment feature importance
]

X = data[features]
y = data["target"]

split = int(len(data) * 0.8)

X_train = X.iloc[:split]
X_test = X.iloc[split:]

y_train = y.iloc[:split]
y_test = y.iloc[split:]
```

```python
# ============ Logistic Regression ===============

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

pred_lr = lr.predict(X_test)

print("Logistic Accuracy:", accuracy_score(y_test, pred_lr))

```

```python
# =============== Random Forest ==================

rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)

pred_rf = rf.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, pred_rf))

# Re-run the feature importance cell after this to see the updated results.
```

```python
# ================= XGBoost =====================

xgb = XGBClassifier(
    n_estimators=300,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="logloss"
)

xgb.fit(X_train, y_train)

pred_xgb = xgb.predict(X_test)

print("XGBoost Accuracy:", accuracy_score(y_test, pred_xgb))

```

```python
# =============== Model Comparison ===============

results = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest", "XGBoost"],
    "Accuracy": [
        accuracy_score(y_test, pred_lr),
        accuracy_score(y_test, pred_rf),
        accuracy_score(y_test, pred_xgb)
    ]
})

results

```

```python
# ============ Feature Importance =================

importances = pd.Series(rf.feature_importances_, index=features)
importances.sort_values().plot(kind="barh")
plt.title("Random Forest Feature Importance (AAPL)")
plt.show()
```

```python
# %%
# ================= RESULTS & INSIGHTS =================

from sklearn.metrics import confusion_matrix
import seaborn as sns

print("\n========== MODEL PERFORMANCE ==========\n")

print("Logistic Regression Report:\n")
print(classification_report(y_test, pred_lr))

print("\nRandom Forest Report:\n")
print(classification_report(y_test, pred_rf))

print("\nXGBoost Report:\n")
print(classification_report(y_test, pred_xgb))


# Accuracy table
results = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest", "XGBoost"],
    "Accuracy": [
        accuracy_score(y_test, pred_lr),
        accuracy_score(y_test, pred_rf),
        accuracy_score(y_test, pred_xgb)
    ]
})

display(results)


# Best model
best_model = results.sort_values("Accuracy", ascending=False).iloc[0]
print(f"\nBest Model: {best_model['Model']} with Accuracy: {best_model['Accuracy']:.3f}")


# Confusion Matrices
models = {
    "Logistic Regression": pred_lr,
    "Random Forest": pred_rf,
    "XGBoost": pred_xgb
}

for name, preds in models.items():
    cm = confusion_matrix(y_test, preds)
    plt.figure()
    sns.heatmap(cm, annot=True, fmt="d")
    plt.title(name)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()


# Feature Importance Interpretation (Random Forest)
fi = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=False)

print("\nRandom Forest Feature Importance:\n")
display(fi)


print("""
================ INSIGHTS ================

1. FinBERT sentiment typically ranks higher than VADER → domain-specific NLP matters.

2. Tree models (RF / XGBoost) outperform Logistic Regression,
   indicating nonlinear relationships between sentiment and returns.

3. news_count contributes meaningfully → volume of news proxies uncertainty.

4. return feature is strong → market momentum still dominates sentiment.

5. Overall accuracy ~55–60% is realistic due to market efficiency.

Limitations:
- Small RSS sample
- No technical indicators
- No lagged sentiment
- No portfolio backtest

Improvements:
- Add RSI / MACD / Moving Averages
- Lag sentiment (t-1, t-2)
- Walk-forward validation
- Strategy backtesting
- Multiple news sources

========================================
""")
```
